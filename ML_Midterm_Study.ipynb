```r
link <- "https://raw.githubusercontent.com/jahn2178/ma322/main/academics.csv"
gpa <- read.csv(link, stringsAsFactors=TRUE)
str(gpa)
```

```r
#1 What is the Target feature here and what type of feature is it? (two words)
```

**Answer**: SAT, numerical

```r
#2 What is the Independent feature and what type of feature is it? (two words)
```

**Answer**: GPA, numerical

```r
#3 What specific technique can we use to achieve our aim? (be specific and as brief as possible)
```

**Answer**: Because there are two quantitative variables we can use, 'Simple Linear Regression (SLR)' is the most suitable technique.

```r
#4 Run the appropriate model in R. What is the estimated slope coefficient? (round to 3 decimals)

model <- lm(SAT ~ GPA, data=gpa)
summary(model)
```

**Answer**: 193.167

```r
#5 Use the model from the previous step to predict the performance of a student who scored 600 on the SAT: (round to 3 decimal places)

predict(model, data.frame(SAT=600))
```

**Answer**: proportion of the variation = multiple R-squared. 81.24%

1. What is the Target feature and what type of feature is it? (two words)

3. What specific technique can we use to achieve our aim? (be as brief as possible)
4. Run the appropriate model in R and state the estimated partial slope coefficient for 'cyli'. (round to 2 decimal places)
5. Use the model from the previous step to predict the highway fuel consumption (hfuel) for a car that has 6 cyli, 3.5 disp, and has size "midsize."
(round to 2 decimal places)
6. What percentage of the variation in 'hfuel' is explained by the model you created? (express as percentage and round to 2 decimal places)
7. What baseline did R choose for 'size', while creating the model? (one word)

```r
site <- "https://raw.githubusercontent.com/jahn2178/ma322/main/epa.csv"
epa <- read.csv(site, stringsAsFactors=TRUE)
str(epa)
```

```r
#1 What is the Target feature and what type of feature is it? (two words)
```

**Answer**: hfuel, quantitative

```r
#2 What are the Independent feature(s)?
```

**Answer**: manufacturer(nominal) model(nominal) disp(numerical) cyli(numerical) size(nominal)

```r
#3 What specific technique can we use to achieve our aim? (be as brief as possible)
```

**Answer**: Multiple Regression since there are more than 1 independent (predictor) variables.

```r
#4 Run the appropriate model in R and state the estimated partial slope coefficient for 'cyli'. (round to 2 decimal places)

model2 <- lm(hfuel ~ cyli, data=epa)
summary(model2)
```

**Answer**: -2.82

```r
#5 Use the model from the previous step to predict the highway fuel consumption (hfuel) for a car that has 6 cyli, 3.5 disp, and has size "midsize." (round to 2 decimal places)

predict(model2, data.frame(cyli=6, disp=3.5, size="midsize"))
```

**Answer**: 23.13

```r
#6 What percentage of the variation in 'hfuel' is explained by the model you created? (express as percentage and round to 2 decimal places)

model3 <- lm(hfuel ~ manufacturer + model + disp + cyli + size, data=epa)
summary(model3)
```

**Answer**: 85.01%

```r
#7 What baseline did R choose for 'size', while creating the model? (one word)
```

**Answer**: mean

```r
#8 How many dummies did R create for 'size', while creating the model?
```

**Answer**: 2
